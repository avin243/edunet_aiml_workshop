```python
climate_data=[
   http://localhost:8889/tree {"city":"city A", "temperature":12, "carbon_footprint":156},
    {"city":"city B", "temperature":45, "carbon_footprint":148},
    {"city":"city C", "temperature":24, "carbon_footprint":150},
    {"city":"city D", "temperature":26, "carbon_footprint":112},
    {"city":"city E", "temperature":52, "carbon_footprint":111}
]
high_temp_threshhold=26
high_temp_cities= [city for city in climate_data if city["temperature"]>high_temp_threshhold]
print("Cities with high temperature: ")
for city in high_temp_cities:
    print(f"{city["city"]} - {city["temperature"]}")
```

```python
sustainability_threshhold = 126
sustanable_cities = list(filter(lambda city: city["carbon_footprint"] < sustainability_threshhold,climate_data))
print("Sustanable cities carbon_footprint: ")
for city in sustanable_cities:
    print(f"{city["city"]} - {city["carbon_footprint"]}")
```

```python
highest_footprint_city=max(climate_data,key=lambda city: city["carbon_footprint"])
print(f"\nCity with the highest carbon footprint:")
print(f"{highest_footprint_city["city"]} - {highest_footprint_city["carbon_footprint"]} Kg CO2")
```
